```python
from sklearn.impute import KNNImputer
imputer=KNNImputer(n_neighbors=12,weights='distance',)
numerical_columns=imputer.fit_transform(numerical_columns) #KNN imputer to fill missing values in numerical columns
```

```python
numerical_columns=pd.DataFrame(numerical_columns)
```

```python
numerical_columns.columns=['age', 'ball_controlling_skills', 'body_reflexes', 'body_balance',
       'jumping_skills', 'penalties_conversion_rate', 'mental_strength',
       'goalkeeping_skills', 'defending_skills', 'passing_skills',
       'dribbling_skills', 'shot_accuracy', 'body_strength_stamina',
       'max_running_speed', 'behaviour_rating', 'matches_played',
       'fitness_rating', 'trophies_won', 'years_of_experience',
       'no_of_disqualifications', 'selection']
```

```python
data = pd.concat([categorical_columns, numerical_columns], axis=1)
```

```python
data
```

```python
data.isna().sum()
```

```python
sns.boxplot(data[['age']]) #handling outliers in age column
```

```python
data=data[data.age<=38]
```

```python
data.shape
```

```python
data['weight'] = data['weight'].str.extract('(\d+)').astype(int) #changing the datatype of weight feature
```

```python
data.weight
```

```python
data['coaching'] = data['coaching'].str.extract('(\d+)').astype(int) #changing the datatype of coaching feature
```

```python
def convert_height(height_str):
    feet, inches = map(int, height_str.split("'"))
    total_inches = feet * 12 + inches
    return total_inches * 2.54
```

```python
data['height'] = data['height'].apply(convert_height) #changing the datatype of height feature and converting into cms
```

```python
data.info()
```

```python
#Encoding
```

```python
data.country.nunique()
```

```python
data=data.drop('country',axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['gender']=le.fit_transform(data['gender'])
data['strong_foot']=le.fit_transform(data['strong_foot'])
data['dedication_level']=le.fit_transform(data['dedication_level'])
```

```python
#Correlation
```

```python
plt.figure(figsize=(10,8))
corr_matrix=data.corr()
sns.heatmap(corr_matrix,annot=True,cmap='YlGnBu')
```

```python
#Feature Scaling
```

```python
x=data.drop('selection',axis=1)
```

```python
y=data['selection']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax=MinMaxScaler()
```

```python
x=minmax.fit_transform(x)
```

```python
x=pd.DataFrame(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,shuffle=True)
```

```python
#Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(solver='newton-cg', class_weight='balanced')
lr_clf=lr.fit(x_train,y_train)
```

```python
y_predictions=lr_clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score,confusion_matrix
```

```python
print('Accuracy=',accuracy_score(y_test,y_predictions))
print('Precision=',precision_score(y_test,y_predictions,average='weighted'))
print('recall_score=',recall_score(y_test,y_predictions,average='weighted'))
print('f1_score=',f1_score(y_test,y_predictions,average='weighted'))
```

```python
confusion_matrix(y_test,y_predictions)
```

```python
from sklearn.model_selection import KFold
Kfold_validator=KFold(5)
```

```python
from sklearn.model_selection import cross_val_score
cv_result=cross_val_score(lr_clf,x,y,cv=Kfold_validator)
cv_result
```

```python
np.mean(cv_result)
```

```python
#K-Nearest Neighbors
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
metric_k=[]
neighbors=np.arange(3,15)
```

```python
for k in neighbors:
    classifier=KNeighborsClassifier(n_neighbors=k,metric='euclidean')
    classifier.fit(x_train,y_train)
    y_predictions=classifier.predict(x_test)
    acc=accuracy_score(y_test,y_predictions)
    metric_k.append(acc)
```

```python
plt.plot(neighbors,metric_k,'o-')
plt.ylabel('Accuracy')
plt.xlabel('k')
plt.grid()
```

```python
classifier=KNeighborsClassifier(n_neighbors=8,metric='euclidean')
classifier.fit(x_train,y_train)
y_predictions=classifier.predict(x_test)
```

```python
print('Accuracy=',accuracy_score(y_test,y_predictions))
print('Precision=',precision_score(y_test,y_predictions,average='weighted'))
print('recall_score=',recall_score(y_test,y_predictions,average='weighted'))
print('f1_score=',f1_score(y_test,y_predictions,average='weighted'))
confusion_matrix(y_test,y_predictions)
```

```python
cv_result=cross_val_score(classifier,x,y,cv=Kfold_validator)
cv_result
```

```python
np.mean(cv_result)
```
